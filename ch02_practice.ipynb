```python
# 2.2.1 데이터 리더
from llama_index.core import SimpleDirectoryReader
```

```python
documents = SimpleDirectoryReader("sample_docs").load_data()
```

```python
for i, document in enumerate(documents):
    print(f"Document {i+1}:")
    print(document.text)
    print("\n")
```

```python
documents = SimpleDirectoryReader("sample_docs", recursive=True).load_data()
```

```python
for i, document in enumerate(documents):
    print(f"Document {i+1}:")
    print(document.text)
    print("\n")
```

```python
# 특정 파일 형식만 로딩하기 P.42
documents = SimpleDirectoryReader(
    "sample_docs", 
    required_exts=[".txt", ".pdf"],
    recursive=True
).load_data()
```

```python
for i, document in enumerate(documents):
    print(f"Document {i+1}:")
    print(document.text)
    print("\n")
```

```python
# 문서 객체의 메타데이터 확인하기
import pprint
for i, document in enumerate(documents):
    print(f"Document {i+1} metadata:")
    pprint.pprint(document.metadata)
    print("\n")
```
